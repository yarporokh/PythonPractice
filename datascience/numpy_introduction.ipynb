```python
from PIL import Image
from IPython.display import display

im = Image.open('flower.jpg')
display(im)
```

```python
array = np.array(im)
print(array.shape)
array
```

```python
mask = np.full(array.shape, 255)
mask
```

```python
mod_array = array - mask
mod_array= mod_array * -1
mod_array = mod_array.astype(np.uint8)
mod_array
```

```python
display(Image.fromarray(mod_array))
```

```python
reshaped = np.reshape(mod_array, (250, 3, 250))
print(reshaped.shape)
```
